```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from seaborn import load_dataset
data = pd.read_csv("Titanic1.csv")
tips = load_dataset("tips")
```

```python
sns.countplot(data=data, x='survived')
plt.show()
```

```python
data['sex'].value_counts().plot(kind="pie", autopct="%.2f")
plt.show()
```

```python
plt.hist(data['age'], bins=5)
plt.show()
```

```python
sns.distplot(data['age'])
plt.show()
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue=tips["sex"])
plt.show()
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue=tips["sex"], style=tips['smoker'])
plt.show()
```

```python
sns.barplot(data=data, x="class", y="age")
plt.show()
```

```python
sns.barplot(data=data, x="class", y="fare", hue = data["sex"])
plt.show()
```

```python
sns.boxplot(data=data, x="sex", y="age")
```

```python
sns.boxplot(data=data, x="class", y="age", hue="survived")
plt.show()
```

```python
sns.distplot(data[data['survived'] == 0]['age'], hist=False, color="blue")
sns.distplot(data[data['survived'] == 1]['age'], hist=False, color="orange")
plt.show()
```

```python
pd.crosstab(data['class'], data['survived'])
```

```python
sns.heatmap(pd.crosstab(data['class'], data['survived']))
```

```python
sns.clustermap(pd.crosstab(data['parch'], data['survived']))
plt.show()
```
